```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Cargar datos
df = pd.read_csv('/content/poblacion_mexico.csv')

#Asegurar que el periodo sea numerico
df['Periodo'] = pd.to_numeric(df['Periodo'], errors='coerce')

#Eliminar filas con valores no numericos
df = df.dropna(subset=['Periodo'])

#Graficar poblacion total
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Periodo', y='Poblacion', marker='o')
plt.title('Crecimiento de la poblacin total en Mxico')
plt.xlabel('Periodo')
plt.ylabel('Poblacion Total')
plt.grid(True)
plt.tight_layout()
plt.show()

#Comparacion hombres vs mujeres
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Periodo', y='Hombres', label='Hombres', marker='o')
sns.lineplot(data=df, x='Periodo', y='Mujeres', label='Mujeres', marker='o')
plt.title('Distribucin por sexo en Mxico')
plt.xlabel('Periodo')
plt.ylabel('Poblacion')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
